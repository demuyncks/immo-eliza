```python
# Introducing regions:  matching zip to region
def classify_region(zipcode):
    prefix = int(str(zipcode)[:2])

    if 10 <= prefix <= 12:
        return "Brussels"
    elif 15 <= prefix <= 39 or 80 <= prefix <= 99:
        return "Flanders"
    elif 40 <= prefix <= 79 or 13 <= prefix <= 14:
        return "Wallonia"
    else:
        return "Unknown"
        
df2['region'] = df2['zip_code'].apply(classify_region)
df2['region'] = df2['region'].astype('str')
```

```python
# Hierarchical imputation for 'geo_smart_index' (based on ZIP, Province, Region, Global)
import pandas as pd

def compute_geo_smart_index(X_train, X_test, y_train):
    """
    Создает колонку 'geo_smart_index' на основе средних цен из y_train.
    Использует иерархию: ZIP -> Province -> Region -> Global.
    """
    # Временный датафрейм для расчета групп (только Train!)
    temp_train = X_train.copy()
    temp_train['target_price'] = y_train

    # 1. Справочники средних цен (Расчет только по TRAIN)
    global_mean = y_train.mean()
    region_map = temp_train.groupby('region')['target_price'].mean()
    prov_map = temp_train.groupby('province')['target_price'].mean()
    zip_map = temp_train.groupby('zip_code')['target_price'].mean()
    
    # 2. Статистика количества продаж по ZIP в Train
    zip_counts = temp_train['zip_code'].value_counts()

    def transform_df(df):
        # А) Создаем "слои" цен для каждой строки
        # .map() сопоставляет текст (напр. 'Flanders') с числом (напр. 410000)
        val_region = df['region'].map(region_map).fillna(global_mean)
        val_prov = df['province'].map(prov_map).fillna(val_region)
        val_zip = df['zip_code'].map(zip_map).fillna(val_prov)
        
        # Б) Проверяем, сколько раз этот ZIP встречался в обучении
        counts = df['zip_code'].map(zip_counts).fillna(0)
        
        # В) ЛОГИКА "МАТРЕШКИ":
        # Если в обучении было >= 5 примеров для этого ZIP — берем его среднее.
        # Если меньше — берем среднее по Провинции.
        smart_index = val_zip.where(counts >= 5, val_prov)
        
        return smart_index

    # 3. Применяем трансформацию к обоим наборам
    train_geo_idx = transform_df(X_train)
    test_geo_idx = transform_df(X_test)
    
    return train_geo_idx, test_geo_idx

# --- ПРАВИЛЬНОЕ ПРИМЕНЕНИЕ В ТВОЕМ КОДЕ ---
# X_train['geo_smart_index'], X_test['geo_smart_index'] = compute_geo_smart_index(X_train, X_test, y_train)
```

```python
import pandas as pd
import numpy as np

def fill_missing_hierarchical(X_train, X_test, columns_to_fill, method='median'):
    """
    Заполняет пропуски в X_train и X_test на основе статистики из X_train.
    Иерархия: ZIP (если >= 5 записей) -> Province -> Region -> Global.
    """
    X_train_filled = X_train.copy()
    X_test_filled = X_test.copy()
    
    # Словари для хранения статистики (отчеты)
    reports = {'global': {}, 'region': {}, 'province': {}, 'zip': {}}
    
    for col in columns_to_fill:
        # 1. Считаем базовые метрики (пока в float64 для безопасности)
        if method == 'median':
            global_val = X_train[col].median()
            # Округляем сразу результат агрегации
            agg_func = lambda x: x.median().round()
        else:
            raw_mode = X_train[col].mode()
            global_val = raw_mode.iloc[0] if not raw_mode.empty else 0
            agg_func = lambda x: x.mode().iloc[0] if not x.mode().empty else None

        # 2. Формируем справочники-матрешки (округляем и храним как float/int)
        reports['global'][col] = round(global_val) if method == 'median' else global_val
        reports['region'][col] = X_train.groupby('region')[col].agg(agg_func)
        reports['province'][col] = X_train.groupby('province')[col].agg(agg_func)
        reports['zip'][col] = X_train.groupby('zip_code')[col].agg(agg_func)
        
        # Справочник частоты для ZIP
        zip_counts = X_train['zip_code'].value_counts()

        def transform_col(df, column_name):
            # Маппинг уровней
            val_region = df['region'].map(reports['region'][column_name]).fillna(reports['global'][column_name])
            val_prov = df['province'].map(reports['province'][column_name]).fillna(val_region)
            val_zip = df['zip_code'].map(reports['zip'][column_name]).fillna(val_prov)
            
            # Проверка порога доверия
            counts = df['zip_code'].map(zip_counts).fillna(0)
            substitute_values = val_zip.where(counts >= 5, val_prov)
            
            # ЗАПОЛНЕНИЕ: сначала fillna, потом приведение к Int64
            # Мы используем .astype(float) перед round, чтобы избежать NAType error
            return df[column_name].fillna(substitute_values).astype(float).round().astype('Int64')

        # Применяем трансформацию
        X_train_filled[col] = transform_col(X_train, col)
        X_test_filled[col] = transform_col(X_test, col)
        
    return X_train_filled, X_test_filled, reports
```

```python
# 1. Список колонок для медианы (площади)
cols_median = ['living_area', 'surface_of_the_land', 'terrace_area', 'garden_area']

# 2. Список колонок для моды (дискретные величины)
cols_mode = ['number_of_rooms', 'number_of_facades', 'state_of_the_building']

# Применяем медианное заполнение
X_train, X_test = fill_missing_hierarchical(X_train, X_test, cols_median, method='median')

# Применяем модальное заполнение
X_train, X_test = fill_missing_hierarchical(X_train, X_test, cols_mode, method='mode')

# И ТОЛЬКО ПОСЛЕ ЭТОГО создаем гео-индекс
X_train['geo_smart_index'], X_test['geo_smart_index'] = compute_geo_smart_index(X_train, X_test, y_train)
```

```python
# Updated version of the function to fill missing values with hierarchical imputation and return reports
import pandas as pd
import numpy as np

def fill_missing_hierarchical(df, stats_report):
    """
    Fills missing values using a hierarchical approach: ZIP -> Province -> Region -> Global.
    Differentiates between property types (HOUSE vs APARTMENT) for better accuracy.
    """
    # List of columns to fill
    target_columns = ['living_area', 'terrace_area', 'garden_area', 'surface_of_the_land', 'number_of_facades']
    
    for col in target_columns:
        if col not in df.columns:
            continue
            
        # We process each row where the target column is NaN
        missing_mask = df[col].isna()
        
        for idx, row in df[missing_mask].iterrows():
            z_code = row['zip_code']
            prov = row['province']
            reg = row['region']
            prop_type = row['property_type']
            
            fill_value = None
            
            # 1. Try to get median from ZIP code (most specific)
            try:
                # We assume stats_report['zip'][col] is a Series indexed by zip_code
                val = stats_report['zip'][col].get(z_code)
                if pd.notna(val):
                    fill_value = val
            except (KeyError, AttributeError):
                pass
            
            # 2. If still NaN, try Province level
            if pd.isna(fill_value):
                try:
                    val = stats_report['province'][col].get(prov)
                    if pd.notna(val):
                        fill_value = val
                except (KeyError, AttributeError):
                    pass
            
            # 3. If still NaN, try Region level
            if pd.isna(fill_value):
                try:
                    val = stats_report['region'][col].get(reg)
                    if pd.notna(val):
                        fill_value = val
                except (KeyError, AttributeError):
                    pass
            
            # 4. Final fallback: Global median (least specific)
            if pd.isna(fill_value):
                fill_value = stats_report['global'].get(col)
            
            # Apply the found value to the dataframe
            df.at[idx, col] = fill_value
            
    return df
```
