```python
from bs4 import MarkupResemblesLocatorWarning
import warnings
warnings.filterwarnings("ignore", category=MarkupResemblesLocatorWarning)


import requests
from bs4 import BeautifulSoup

#get cookies
#url_cookies="https://immovlan.be/fr/cookies"
#r=requests.get(url_cookies)
#cookies=r.cookies

#url="https://immovlan.be/fr/projectdetail/1951886-7574316"

urls=["https://immovlan.be/en/projectdetail/1951886-7574316",
"https://immovlan.be/en/detail/appartement/a-vendre/1030/schaerbeek/vbd88427",
"https://immovlan.be/en/detail/immeuble-mixte/a-vendre/1030/schaerbeek/vbd31200",
"https://immovlan.be/en/projectdetail/2573463-7554937",
"https://immovlan.be/en/detail/immeuble-mixte/a-vendre/1030/schaerbeek/vbd89569",
"https://immovlan.be/en/detail/surface-industrielle/a-vendre/1030/schaerbeek/vbd89703"]

headers={"User-Agent":"Mozilla/5.0 (Linux; Android 10; K) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/143.0.0.0 Mobile Safari/537.36"}

#Create list of 7 webpages
list_html=[]
for url in urls:
    r=requests.get(url,headers=headers)
    html_text=r.text
    list_html.append(html_text)

list_html
```

```python
#Find pattern in html for main properties
count=0

for html in list_html:
    soup=BeautifulSoup(html,"html.parser")
#soup.find_all("strong")
    data=soup.find_all(class_="property-highlight margin-bottom-05 margin-right-05")
    print(f"Property number {count+1} at {urls[count]}")
    count+=1
    print(data)
```

```python
#Find pattern in html for general informations
count=0

for html in list_html:
    soup=BeautifulSoup(html,"html.parser")
#soup.find_all("strong")
    general_info=soup.find_all(class_="general-info-wrapper")
    count+=1
    print(f"Property number {count} at {urls[count-1]}")

    for infos in general_info:
        specific_info=infos.find_all("h4")
        specific_value=infos.find_all("p")
        for index, value in enumerate(specific_info):
            print(f"{specific_info[index].get_text()} : {specific_value[index].get_text()}")
            index+=1
```
