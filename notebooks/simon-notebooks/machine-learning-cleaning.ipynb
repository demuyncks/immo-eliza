```python
# Load Dataset into X and y
path='../../data/cleaned_properties_dataset_2026-03-12-13h12.csv'
df=pd.read_csv(path)
df=df.drop(['url','zip_code','city','price_sqm'],axis=1)

df=pd.get_dummies(df)

X=df.drop(['price'],axis=1).to_numpy()
y=df['price'].to_numpy()
```

```python
# Split into train and test and process
def split_clean_data(X,y,test_size):
    X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=test_size,random_state=42)

    # Replace missing values in X_train and X_test
    ## X_train
    df=pd.DataFrame(X_train)
    df_num = df.select_dtypes(include=['int64', 'float64'])
    df_num = df_num.fillna(df_num.median())
    df_bool=pd.DataFrame(X_train)
    df_bool = df.select_dtypes(include='bool')
    df_bool = df_bool.fillna(df_bool.mode())
    df_str = df.select_dtypes(include='str')
    X_train = pd.concat([df_num, df_bool,df_str], axis=1).to_numpy()

    ## X_test
    df=pd.DataFrame(X_test)
    df_num = df.select_dtypes(include=['int64', 'float64'])
    df_num = df_num.fillna(df_num.median())
    df_bool=pd.DataFrame(X_test)
    df_bool = df.select_dtypes(include='bool')
    df_bool = df_bool.fillna(df_bool.mode())
    df_str = df.select_dtypes(include='str')
    X_test = pd.concat([df_num, df_bool,df_str], axis=1).to_numpy()

    return X_train,X_test,y_train,y_test
```
