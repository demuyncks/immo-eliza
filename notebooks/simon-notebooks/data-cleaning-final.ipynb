```python
import pandas as pd
import numpy as np
```

```python
# 01- Import Dataset

path="../../data/cleaned_properties_dataset_2026-03-12-13h12.csv"
properties=pd.read_csv(path,index_col=0)
```

```python
#02 - First Cleaning

empty_lines_price=properties[properties['price'].isna() == True].index
empty_lines_area=properties[properties['living_area'].isna()==True].index
empty_lines=empty_lines_price.append(empty_lines_area)

# Remove empty lines
properties=properties.drop(empty_lines,axis=0)

# Drop Terrace
properties.drop(['terrace_area'],axis=1,inplace=True)
properties[['open_fire','furnished','garden','swimming pool','terrace']]=properties[['open_fire','furnished','garden','swimming_pool','terrace']].replace(np.nan,0)

# Remove spaces before and/or after if there are (>> strip)
properties=properties.apply(lambda x: x.str.strip() if x.dtype=="object" else x)

# Correct one specific value
properties.loc[properties.city=='La','city']="La Louvière"

# Garden area = 0 if Garden is False
properties.loc[properties.garden == 0,'garden_area']=0

# Drop prop with price <10000
properties=properties[properties['price']>=10000]
```

```python
# Change types
## to string
properties["zip_code"]=properties["zip_code"].astype(int).astype(str)

## to int
floats_col=properties.select_dtypes('float64').columns
floats_col
properties[floats_col]=properties[floats_col].astype("Int64")
```
