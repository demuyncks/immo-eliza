```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import StandardScaler
import joblib
```

```python
input={
  "number_of_facades": 1,
  "number_of_rooms": 2,
  "open_fire": False,
  "province": "Antwerp",
  "region": "Flanders",
  "state_of_the_building": "To renovate",
  "subtype_of_property": "apartment",
  "surface_of_the_land": 12,
  "swimming_pool": False,
  "terrace": False,
  "terrace_area": np.nan,
  "type_of_property": "Apartment",
  "type_of_sale": "in public sale",
  "zip_code": "1000"
}
input_df=pd.DataFrame([input])
```
