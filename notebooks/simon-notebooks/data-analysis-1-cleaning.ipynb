# Step 1 : Data Cleaning
#### You have collected your data! So it's time to do a cleaning on it. A cleaned dataset is a dataset that doesn't contain any duplicates, is blank spaces or error-free. The rest of the analysis can be discarded if you neglect this step!

### No duplicates
### No blank spaces (ex: " I love python " => "I love python")
### No errors
### No empty values


```python
# Get stats and infos about dataset
properties.describe()
properties.info()

# Check if duplicates by url --> bool
properties.URL.is_unique

#Check empty lines
empty_lines_price=properties[properties['Price'].isna() == True].index
empty_lines_area=properties[properties['Living Area'].isna()==True].index
empty_lines=empty_lines_price.append(empty_lines_area)
```

Spaces in Name of City causes a problem (La Louvière > La)

We have floats instead of int

Zip is float instead of str

‘Terrace area’ → all NaN OK

Row 1179 no infos > drop: Ok

```python
# Remove empty lines
properties=properties.drop(empty_lines,axis=0)

# Drop Terrace
properties.drop(['Terrace area'],axis=1,inplace=True)
properties[['Open fire','Furnished','Garden','Swimming pool','Terrace']]=properties[['Open fire','Furnished','Garden','Swimming pool','Terrace']].replace(np.nan,0)

# Remove spaces before and/or after if there are (>> strip)
properties=properties.apply(lambda x: x.str.strip() if x.dtype=="object" else x)

# Correct one specific value
properties.loc[properties.City=='La','City']="La Louvière"

# Garden area = 0 if Garden is False
properties.loc[properties.Garden == 0,'Garden area']=0

# Drop prop with price <10000
properties=properties[properties['Price']>=10000]
```

```python
# Change types
## to string
properties["Zip code"]=properties["Zip code"].astype(int).astype(str)

## to int
floats_col=properties.select_dtypes('float64').columns
floats_col
properties[floats_col]=properties[floats_col].astype("Int64")

properties.dtypes
```

```python
# Change column names
properties.columns=['url', 'zip_code', 'city', 'type_of_property',
       'subtype_of_property','price', 'type_of_sale',
       'number_of_rooms', 'living_area', 'fully_equipped_kitchen', 'furnished',
       'open_fire', 'terrace', 'garden', 'garden_area',
       'surface_of_the_land', 'number_of_facades', 'swimming_pool',
       'state_of_the_building']
properties
```
