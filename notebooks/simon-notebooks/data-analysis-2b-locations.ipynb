```python
import pandas as pd
import seaborn as sns
import numpy as np
from matplotlib import pyplot as plt
```

```python
# Load csv as df
path="../../data/simon-data/properties-2-extrapolated.csv"
properties=pd.read_csv(path,index_col=0)
properties=properties.reset_index()
properties.drop('index',axis=1)
```

```python
# Assign Province and Region
properties[['Province','Region']]=np.nan

## Province by Zip
properties["Zip code"]=properties["Zip code"].astype(str)
provinces = {
    "1": "Brussels",
    "2": "Antwerp",
    "30": "Flemish Brabant", "31": "Flemish Brabant", "32": "Flemish Brabant",
    "33": "Flemish Brabant", "34": "Flemish Brabant", "35": "Flemish Brabant",
    "36": "Limburg", "37": "Limburg", "38": "Limburg", "39": "Limburg",
    "4": "Liège",
    "5": "Namur",
    "6": "Hainaut",
    "7": "Hainaut",
    "8": "West Flanders",
    "9": "East Flanders"
}

def get_province(x):
    return provinces.get(x[:2]) or provinces.get(x[0])

properties["Province"] = properties["Zip code"].apply(get_province)

## Region by Zip
regions = {
    "1": "Brussels",
    "2": "Flemish Region",
    "3": "Flemish Region",
    "4": "Walloon Region",
    "5": "Walloon Region",
    "6": "Walloon Region",
    "7": "Walloon Region",
    "8": "Flemish Region",
    "9": "Flemish Region"
}
def get_region(x):
    return regions.get(x[0])

properties["Region"]=properties["Zip code"].apply(get_region)
```

```python
# Save csv
properties.to_csv("../../data/simon-data/properties-2b-located.csv")
```
