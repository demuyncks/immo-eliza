```python
import pandas as pd
import seaborn as sns
import numpy as np
from matplotlib import pyplot as plt
```

```python
# Load csv as df
path="../../data/simon-data/properties-2-extrapolated.csv"
properties=pd.read_csv(path,index_col=0)
properties=properties.reset_index()
properties.drop('index',axis=1)
properties["Zip code"] = properties["Zip code"].astype(int)
```

```python
# Load provinces Dataset
path="../../data/simon-data/provinces-zip.csv"
provinces_zip=pd.read_csv(path)
provinces_zip=provinces_zip[['zipCode','province']]
provinces_zip=provinces_zip.drop(0)[['zipCode','province']].set_index('zipCode')
provinces_zip=provinces_zip.to_dict()
```

```python
provinces_zip
```

```python
# Assign Province and Region
properties[['Province','Region']]=np.nan

## Province by Zip
translation = {
    "Limburg": "Limburg",
    "Luxemburg": "Luxembourg",
    "Oost-Vlaanderen": "East Flanders",
    "West-Vlaanderen": "West Flanders",
    "Vlaams-Brabant": "Flemish Brabant",
    "Antwerpen": "Antwerp",
    "Luik": "Liège",
    "Namen": "Namur",
    "Henegouwen": "Hainaut",
    "Waals-Brabant": "Walloon Brabant",
    "Brussel": "Brussels"
}

# Appliquer la traduction sur les valeurs du dict
zip_to_province = {str(k): translation[v] for k, v in provinces_zip["province"].items()}

properties["Zip code"]=properties["Zip code"].astype(str)
properties["Province"] = properties["Zip code"].map(zip_to_province)


## Region by Zip
province_to_region = {
    "Brussels": "Brussels",
    "Antwerp": "Flemish Region",
    "East Flanders": "Flemish Region",
    "West Flanders": "Flemish Region",
    "Flemish Brabant": "Flemish Region",
    "Limburg": "Flemish Region",
    "Liège": "Walloon Region",
    "Namur": "Walloon Region",
    "Hainaut": "Walloon Region",
    "Walloon Brabant": "Walloon Region",
    "Luxembourg": "Walloon Region"
}

properties["Region"]=properties["Province"].map(province_to_region)
```

```python
# Save csv
properties.to_csv("../../data/simon-data/properties-2b-located.csv")
```
