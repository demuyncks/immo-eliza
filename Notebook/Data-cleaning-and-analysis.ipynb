```python
# 1. Remove exact duplicates
df.drop_duplicates(inplace=True)

# 2. Remove rows missing vital info
df.dropna(subset=['Price', 'Living Area', 'Zip code', 'City'], inplace=True)

# 3. Handle the "Pool" (Assume missing means No Pool)
df['Swimming pool'] = df['Swimming pool'].fillna(0)
df['Open fire'] = df['Open fire'].fillna(0)
df['Terrace'] = df['Terrace'].fillna(0)
df['Garden'] = df['Garden'].fillna(0)

# 4. Create the Price per m2 column
df['Price_per_m2'] = df['Price'] / df['Living Area']
```

```python
# Check how many rows/columns are left
print(f"Remaining rows: {df.shape[0]}")

# Check specifically for nulls in those columns
print(df[['Price', 'Living Area', 'Swimming pool','Zip code', 'City', 'Open fire','Terrace', 'Garden' ]].isnull().sum())
```

```python
# 5. Create a column to categorize the region
def get_region(Zip_code):
    z = int(Zip_code)
    if 1000 <= z <= 1299: 
        return 'Brussels'
    if (1300 <= z <= 1499) or (4000 <= z <= 7999): 
        return 'Wallonia'
    return 'Flanders'

df['Region'] = df['Zip code'].apply(get_region)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Plotting the price distribution to see the outliers
plt.figure(figsize=(10, 6))
sns.boxplot(x=df['Price'])
plt.title('Price Distribution & Outliers')
plt.show()
```

```python
# Average price per square meter by City
geo_analysis_perm2 = df.groupby(['City', 'Region'])['Price_per_m2'].mean().nlargest(5)
print (geo_analysis_perm2)
# Average price by City
geo_analysis = df.groupby(['City', 'Region'])['Price'].mean().nlargest(5)
print (geo_analysis)

# Top 5 most expensive cities in Flanders
flanders_top = df[df['Region'] == 'Flanders'].groupby('City')['Price_per_m2'].nlargest(5)
print (flanders_top)
```

```python
# Correlation between price and living area
correlation = df['Living Area'].corr(df['Price'])
print(f"Correlation: {correlation}")
"""The correlation is not high. It does not show that the larger house would have better price in Belgium. Indeed, location/Condition is King" — people pay more for a small nice house in a good city than a big house in a bad/rural area."""
```

```python
# 'numeric_only=True' is vital so it ignores text columns like City
correlation_matrix = df.corr(numeric_only=True)
correlation_matrix_Flanders = df[df['Region'] == 'Flanders'].corr(numeric_only=True)
correlation_matrix_Brussels = df[df['Region'] == 'Brussels'].corr(numeric_only=True)
correlation_matrix_Wallonia = df[df['Region'] == 'Wallonia'].corr(numeric_only=True)

# To see how everything specifically correlates to PRICE:
price_correlations = correlation_matrix['Price'].sort_values(ascending=False)
price_correlations_Flanders = correlation_matrix_Flanders['Price'].sort_values(ascending=False)
price_correlations_Brussels = correlation_matrix_Brussels['Price'].sort_values(ascending=False)
price_correlations_Wallonia = correlation_matrix_Wallonia['Price'].sort_values(ascending=False)
print(price_correlations)
print(price_correlations_Flanders)
print(price_correlations_Brussels)
print(price_correlations_Wallonia)
```

```python
# Select only your requested columns
cols_to_analyze = [
    "Price", "Number of rooms", "Living Area", "Surface of the land", 
    "Swimming pool", "Number of facades", "Open fire", 
    "State of the building", "Garden", "Terrace", 
    "Furnished", "Fully equipped kitchen"
]

regions = ['Brussels', 'Flanders', 'Wallonia']

# Create a figure with 3 side-by-side plots
fig, axes = plt.subplots(1, 3, figsize=(22, 7))

for i, region in enumerate(regions):
    # Filter data for the region
    region_df = df[df['Region'] == region][cols_to_analyze]
    
    # Calculate correlation
    corr = region_df.corr()
    
    # Draw the heatmap on the specific 'axis'
    sns.heatmap(corr, annot=True, fmt=".2f", cmap='RdBu_r', center=0, 
                ax=axes[i], cbar=(i == 2), vmin=-1, vmax=1)
    
    axes[i].set_title(f'Market Drivers in {region}', fontsize=15)
    axes[i].set_xticklabels(axes[i].get_xticklabels(), rotation=45, ha='right')

plt.suptitle('Comparison of Real Estate Price Drivers by Belgian Region', fontsize=20, y=1.05)
plt.tight_layout()
plt.show()
```

1. Brussels: "The Density King"
The Data: Highest correlation with Living Area (0.80) and Number of rooms (0.64).

The Insight: In a city where space is a luxury, every square meter counts. Buyers are paying primarily for the interior volume.

2. Wallonia: "The Estate & Luxury Market"
The Data: Highest correlation for Surface of the land (0.45) and Swimming pool (0.37) with Price.

The Insight: Wallonia is where you go for lifestyle. Because land is more available, it is a primary driver of the price. The high "Swimming pool" correlation (0.37 vs 0.25 in Brussels) proves that the luxury segment (villas/estates) is a more dominant force in the Walloon price index.

3. Other insights
House furniture and furnished kitchen: In all three regions, Furnish and Fully equipped kitchen has almost zero correlation with price (0.01 to 0.03). => Belgian prefer decorating house on their own, furnished house will not boost the price of the house.
Garden, Terrace, Swimming polll: Note the low correlation for those elements. This suggests that in Belgium, having a garden/terrace/swimming pool is a "bonus" but doesn't systematically drive the total price as much as an extra bedroom does.
